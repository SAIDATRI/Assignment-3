# Data Visulization

```python
##Check the highest and lowest price areas in the Neighbourhood Groups
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(y="price",x ='neighbourhood_group' ,data = df1)
plt.title("neighbourhood_group price distribution")
plt.show()
```

From the above analysis, we can say that the price distribution order of the neighbourhoods are Manhattan > Brooklyn > Queens > Staten Island > Bronx

```python
##Checking About Neighbourhood Groups
```

```python
plt.hist(df1['neighbourhood_group'], bins=20)
plt.title("Neighbourhood Group",size=15, weight='bold')
plt.xlabel("Neighbourhood Group", size=13)
plt.show()
```

More Neighbourhood Groups are present in Manhattan and least groups in Staten Islsnd

```python
## for Room Prefering
```

```python
plt.scatter(df1['room_type'], df1['minimum_nights'], color="blue")
plt.title("Traffic Areas",size=15, weight='bold')
plt.xlabel("Room Type", size=13)
plt.ylabel("Minimum number of nights stayed", size=13)
plt.show()
```

From the above analysis, it is clear that people prefering Entire hone/apt and private room when compared to shared rooms

```python
##highly prefered rooms
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x = 'room_type',hue = "neighbourhood_group",data = df1)
plt.title("Room types occupied by the neighbourhood_group")
plt.show()
```

From the above analysis, it is clear that people prefering Manhattan more

```python
## high neighbourhood_group
```

```python
sns.countplot(df1["neighbourhood_group"])
```

From the above analysis, it is clear that neighbourhood_group are more in Manhattan.

```python
##Busiest neighbourhood
```

```python
fig = plt.figure(figsize =(10, 7))
plt.bar('neighbourhood','number_of_reviews',data=df1,color='red')
plt.title("Busiest Place",size=15, weight='bold')
plt.xlabel("Name of the neighbourhood", size=13)
plt.ylabel("Number of Reviews", size=13)
plt.show()
```

```python
##Density and distribution of prices for each neighberhood_group
```

```python
sns.violinplot(data=df1, x='neighbourhood_group', y='price')
plt.title("Density and distribution of prices for each neighberhood_group",size=15, weight='bold')
```

```python
## For highest stay
```

```python
plt.figure(figsize=(20,15))
sns.set_palette("Set1")

sns.lineplot(x='price', y='number_of_reviews', data=df1[df1['neighbourhood_group']=='Brooklyn'], label='Brooklyn')
sns.lineplot(x='price', y='number_of_reviews', data=df1[df1['neighbourhood_group']=='Manhattan'],label='Manhattan')
sns.lineplot(x='price', y='number_of_reviews', data=df1[df1['neighbourhood_group']=='Queens'], label='Queens')
sns.lineplot(x='price', y='number_of_reviews', data=df1[df1['neighbourhood_group']=='Staten Island'], label='Staten Island')
sns.lineplot(x='price', y='number_of_reviews', data=df1[df1['neighbourhood_group']=='Bronx'], label='Bronx')
plt.title("Price vs Number of Reviews vs Neighbourhood Group",size=15, weight='bold')
plt.xlabel("Price", size=13)
plt.ylabel("Number of Reviews", size=13)
plt.show()
```

from the above analysis, we can say that most people prefer to stay in place where price is less

```python
## Know about price Distribution
```

```python
plt.figure(figsize=(10,10))
sns.distplot(df1['price'])
plt.title("Price Distribution Plot",size=15, weight='bold')
```

From the above analysis, it is clear that the price is distributed between 30-800 and average price will be 150

```python
## Checking Correlation
```

```python
plt.figure(figsize=(10,10))
palette = sns.diverging_palette(20, 220, n=256)
corr=df1.corr(method='pearson')
sns.heatmap(corr, annot=True, fmt=".2f", cmap=palette, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5}).set(ylim=(11, 0))
plt.title("Correlation Matrix",size=15, weight='bold')
plt.show()
```
